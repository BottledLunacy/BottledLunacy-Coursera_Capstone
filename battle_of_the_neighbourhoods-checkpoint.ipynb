```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

### This function will find venues within a radius of each company and add them to the dataset

```python
def getNearbyVenues(names, region, latitudes, longitudes, radius, limit):
    
    venues_list=[]
    for name, region, lat, lng in zip(names, region, latitudes, longitudes):       
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            limit)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(           
            name, 
            region,
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['COMPANYNAME', 
                             'REGION',
                             'Latitude', 
                              'Longitude', 
                              'Venue', 
                              'Venue Latitude', 
                              'Venue Longitude', 
                              'Venue Category']
    
    return(nearby_venues)
```

# 1. Identifying the Business Problem <a name="problem"></a>

FDI (Foreign Direct Investment) in Ireland comprises around __[20%](https://enterprise.gov.ie/en/What-We-Do/Trade-Investment/Foreign-Direct-Investment-FDI-/)__ of all private sector employment with Dublin, Cork, Kildare and Limerick playing host to the top players such as Amazon, Apple, Facebook, Google, Microsoft and  Salesforce being some of the largest employers. The companies selected are based on the __[top1000](http://www.top1000.ie/industries/technology-business)__ some of which have multiple buildings in different locations. This is by no means an exhaustive list but is sufficient to represents the top employers by employee numbers.

<img src="fdi_stats.png" alt="Alt text" title="Title text" style="width:290px;height:329px;"/>

Young top talent coming out of universities in many countries interested securing employment with these companies know that Ireland is a multicultural society and that the Irish are infamous for being friendly and engaging. They to ensure that they select their potential employer based on the location offering a wide variety of social venues within walking distance of the office (2km). 

The objective of this project is discover which company location offers the widest range and variety of social venues within a 2km radius using the Foursquare API. Dublin City has the largest portion of the population and hosts the majority of the selected companies hence would be expected to have the most variety in venues. However, Limerick and Cork are also thriving cities and would be expected to have a decent variety but the key differentiator here will the location of each company. Some companies tend to locate in business parks outside of the city centre which tends to limit venue availability within a 2km range. The distance of 2km is about twenty minute walk.

## 5. Modelling

I will use the supervised learning algorithm k-means clustering to cluster venues around each company.

```python
# one hot encoding
company_onehot = pd.get_dummies(df_venues[['VenueGrouping']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
company_onehot['COMPANYNAME'] = df_venues['COMPANYNAME'] 

# move neighborhood column to the first column
fixed_columns = [company_onehot.columns[-1]] + list(company_onehot.columns[:-1])
companyvenues_onehot = company_onehot[fixed_columns]

companyvenues_onehot.head()
```

Get Dataframe size

```python
companyvenues_onehot.shape
```

#### Next, let's group rows by company and by taking the mean of the frequency of occurrence of each category

```python
df_merged_companyvenues_grouped = company_onehot.groupby('COMPANYNAME').mean().reset_index()
df_merged_companyvenues_grouped.head()
```

I will calculate the optimum k-means K value using the Elbow technique to determine  the lowest distortion non-zero value. There are 11 venue groupings hence that will feature int he kcluster setting

```python
kclusters = (11,21)

dfEblow = df_merged_companyvenues_grouped.drop('COMPANYNAME', 1)

distortions = []

#The range was set at 7 which is the max that can use without the code crashing
K = range(1,20)
for k in K:
    kmean = KMeans(n_clusters=k, random_state=0, n_init = 50, max_iter = 500)
    kmean.fit(dfEblow)
    distortions.append(kmean.inertia_)
```

Visualise the distortions

```python
plt.figure(figsize=(8,5))
plt.plot(K, distortions, 'bx-')
plt.xlabel('K')
plt.ylabel('Distortion')
plt.title('Elbow Method Results')
plt.xticks(rotation=45,horizontalalignment='right', fontweight='light', fontsize='small')
plt.yticks(fontweight='light', fontsize='small',)
plt.show()
```

OptClusterNumber will be used later when running k-means. We want to avoid over-fitting the model and to do we'll look at where the curve sharply changes. There are two points, the first is around 3 K and the second around 6. Initially 6 looked like a good candidate but after running k-means and evaluating the 6th cluster, the dataframe was empty. The maximum number of clusters for which there will be data is 5

Write a function to sort the venues in descending order

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Create the new dataframe and display the top 10 venues for each neighborhood

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd', 'th']

# create columns according to number of top venues
columns = ['COMPANYNAME']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
df_merged_companyvenues_sorted = pd.DataFrame(columns=columns)
df_merged_companyvenues_sorted['COMPANYNAME'] = df_merged_companyvenues_grouped['COMPANYNAME']

for ind in np.arange(df_merged_companyvenues_grouped.shape[0]):
    df_merged_companyvenues_sorted.iloc[ind, 1:] = return_most_common_venues(df_merged_companyvenues_grouped.iloc[ind, :], num_top_venues)

df_merged_companyvenues_sorted.head()
```

### Visualize the resulting clusters around each company

I will use Folium to display the companies and the clusters around them. The companies will be displated as Markers with text so that, when you zoom in, you can see the company name without having to click on the pin. I also draw a circle around each company with a 2000m radius. The idea is, that when you zoom in on a company, you can easily see which clusters are within or outside of that walking radius. Finally, when I display the clusters I'll set their colours using range defined by the number of clusters and then populate a colour coded legend with the cluster name and associated venues which that has the same colours as cluster markers. 

The cluster charts offer some insight into the venue category frequencies by country and company but what they don't show is how both the clusters and venue categories sit within or outside of a 2km radius of each company. I used Folium maps creating text markers for each company (save having to click on the pin for the company name) and surrounded each company with circle of 2k metres. The clusters were added as coloured markers along with legend using the same colours which shows the cluster and associated venue groupings.

The choice of companies distributed in different counties is intention but it also makes displaying their results space consuming on the blog. So I apologies in advance for all of the scrolling you're about to do. That said, visualising the results by region is far more effective than showing datasets with large volumes of rows. I'll break out the visualisation into county and area. As you work through the visuals, a number of patterns begin to emerge:

Where a company is located in a business park, the range of venues diminishes significantly when compared to an urban area.
Outside of the city centres, venues don't generally congregate around the companies. Instead they spread out to the outer boundaries of the radius limit indicating a longer walking times.

```python
import matplotlib.cm as cm
import random

VenueFeatureGroups = []
for venue in dublin_merged['VenueGrouping'].unique():
    VenueFeatureGroups.append(dict({'Name':venue, 'FG':folium.FeatureGroup(name=venue)}))
    

latitude = dfCompanies.loc[4, 'COMPANYLAT'] # neighborhood latitude value
longitude = dfCompanies.loc[4, 'COMPANYLNG'] # neighborhood longitude value

map_clusters = folium.Map(location=[latitude, longitude], zoom_start=7, tiles='cartodbpositron')

#set colour scheme for clusters and legend
number_of_colors = OptClusterNumber
spectral = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
             for i in range(number_of_colors)]

#display the companies so the user can look at the clustering in relation to their location
for lat, lon, name in zip(dfCompanies['COMPANYLAT'],dfCompanies['COMPANYLNG'],dfCompanies['COMPANYNAME']):
    label = f'{name}'
    label = label.replace("International", '')
    label = folium.Popup(label, parse_html=True)
    feature_group = folium.FeatureGroup(f'{name}')
    
    folium.Marker(location=[lat,lon], icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(27,12),
            html='<div style="font-size: 10pt; color : black">%s</div>' % (name.replace("International", '')),
            )).add_to(map_clusters)    

    folium.Circle(location=[lat,lon],radius=2000, color='green').add_to(map_clusters)
   
markers_colors = []
for lat, lon, types, cluster in zip(dublin_merged['Venue Latitude'], dublin_merged['Venue Longitude'], dublin_merged['VenueGrouping'], dublin_merged['Cluster Labels']):
    
    fg = [d['FG'] for d in VenueFeatureGroups if d['Name']==types][0]   
    
    label = folium.Popup(str(types) + ' Cluster ' + str(cluster), parse_html=True)
    
    fg.add_child(folium.CircleMarker(
        [lat, lon],
        radius=0,
        popup=label,        
        fill=True,
        fill_color=spectral[cluster],
        fill_opacity=1))
    map_clusters.add_child(fg)

folium.LayerControl().add_to(map_clusters)

    
#https://nbviewer.jupyter.org/gist/talbertc-usgs/18f8901fc98f109f2b71156cf3ac81cd
template = """
{% macro html(this, kwargs) %}

<!doctype html>
<html lang="en">
<head>
  <meta charset="utf-8">
  <meta name="viewport" content="width=device-width, initial-scale=1">
  <title>jQuery UI Draggable - Default functionality</title>
  <link rel="stylesheet" href="//code.jquery.com/ui/1.12.1/themes/base/jquery-ui.css">

  <script src="https://code.jquery.com/jquery-1.12.4.js"></script>
  <script src="https://code.jquery.com/ui/1.12.1/jquery-ui.js"></script>
  
  <script>
  $( function() {
    $( "#maplegend" ).draggable({
                    start: function (event, ui) {
                        $(this).css({
                            right: "auto",
                            top: "auto",
                            bottom: "auto"
                        });
                    }
                });
});

  </script>
</head>
<body>

 
<div id='maplegend' class='maplegend' 
    style='position: absolute; z-index:9999; border:2px solid grey; background-color:rgba(255, 255, 255, 0.8);
     border-radius:6px; padding: 10px; font-size:14px; right: 20px; bottom: 20px;'>
     
<div class='legend-title'>Legend</div>
<div class='legend-scale'>
  <ul class='legend-labels'>
  """

#in the folium label, I want to display the top 3 venue types for each category as a useful visual shortcut. Any more than that the label becomes very long.
#Learnt how to dynamically build a legend from https://github.com/mrcagney/examples_folium/blob/develop/notebooks/categorical_legend.ipynb for dynamic labels
#display the cluster number and unique venue groupings

#get cluster numbers
for cls in list(dublin_merged['Cluster Labels'].unique()):
    venuestr = ''
    vstr = ''
    
    #get venue groupings for each cluster
    vl = list(dublin_merged.loc[dublin_merged['Cluster Labels'] == cls, dublin_merged.columns[[0] + list(range(1, dublin_merged.shape[1]))]].VenueGrouping.unique())
    
    shape = dublin_merged.loc[dublin_merged['Cluster Labels'] == cls, dublin_merged.columns[[0] + list(range(1, dublin_merged.shape[1]))]].shape[0]
    for v in vl:
        vstr = vstr + ' ' + v
    
    venuestr = 'Cluster: %s: Size %d: Venues: %s ' % (cls, shape, vstr)
    template += f"<li><span style='background:{spectral[cls]}'></span>{venuestr}</li>"
  

template += """    
  </ul>
</div>
</div>
 
</body>
</html>

<style type='text/css'>
  .maplegend .legend-title {
    text-align: left;
    margin-bottom: 5px;
    font-weight: bold;
    font-size: 90%;
    }
  .maplegend .legend-scale ul {
    margin: 0;
    margin-bottom: 5px;
    padding: 0;
    float: left;
    list-style: none;
    }
  .maplegend .legend-scale ul li {
    font-size: 80%;
    list-style: none;
    margin-left: 0;
    line-height: 18px;
    margin-bottom: 2px;
    }
  .maplegend ul.legend-labels li span {
    display: block;
    float: left;
    height: 16px;
    width: 30px;
    margin-right: 5px;
    margin-left: 0;
    border: 1px solid #999;
    }
  .maplegend .legend-source {
    font-size: 80%;
    color: #777;
    clear: both;
    }
  .maplegend a {
    color: #777;
    }
</style>
{% endmacro %}"""

macro = MacroElement()
macro._template = Template(template)

map_clusters.get_root().add_child(macro)
map_clusters.save("clusters.html")
map_clusters
```
